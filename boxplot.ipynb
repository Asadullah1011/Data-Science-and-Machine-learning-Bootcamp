```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
emp = {'Age_years': [10,23,22,34,45,34,23,23,56,35,45,126], 'Salary_Thousands': [15,24,34,56,65,34,23,36,45,26,45,26]}
df = pd.DataFrame(data = emp)
df
```

```python
df.plot(kind = 'box', figsize= (8,5), color = 'r', yticks = range(0,55,5))
plt.show()
```

```python
plt.figure(figsize = (5,6))
ax= sns.boxplot(data = df)
plt.yticks(range(5,55,5))

plt.xlabel('Age and Salary')
plt.ylabel('Numbers')
plt.show()
```
