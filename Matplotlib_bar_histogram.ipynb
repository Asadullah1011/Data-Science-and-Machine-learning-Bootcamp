```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dict = {'Indian Runs' : [10,12,45,23,25,20,22], 'England Runs': [4,8,10,9,10,19,22]}
df = pd.DataFrame(dict,index=[1,2,3,4,5,6,7])
df
```

```python
df['Indian Runs'].plot(kind = 'bar')
plt.show()
```

```python
df.plot(kind = 'bar', figsize=(8,6))
plt.xlabel('Overs', color= 'b', fontsize = 15)
plt.ylabel('Runs', color= 'r', fontsize = 15)
plt.title('Runs per over', color = 'r', fontsize = 20)
plt.grid(axis = 'y')
plt.yticks(range(0,25,2))
plt.show()
```
