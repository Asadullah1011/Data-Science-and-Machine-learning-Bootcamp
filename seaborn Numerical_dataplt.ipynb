```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.tail()
```

```python
sns.relplot(x='total_bill', y= 'tip', data = tips)
plt.show()
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x='total_bill', y='tip', data = tips, hue ='day', style = 'time')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', data = tips, hue ='smoker', style = 'day')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y = 'tip', data = tips, hue ='sex', palette = "bright")   #paletter color= muted, deep,bright, pastel, dark., color blind
plt.show()
```

```python
sns.lineplot(x= 'total_bill', y= 'tip', data = tips, hue = 'smoker')
plt.show()
```
